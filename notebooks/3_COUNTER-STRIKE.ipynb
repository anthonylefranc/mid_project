```python
import pandas as pd
import numpy as np
import re
import cleaning_functions as cfun
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_ransomware = pd.read_csv('..\data\cleaned\clean_dataset.csv')
```

```python
# In this section, we will try to understand the type of reactions businesses have when attacked by a Ransomware, 
# do companies usually pay or prefere not to ?

# We will check over time what the behavior of companies is, to pay or refuse to pay
# Then we will check the number of paid ransom per countries, sectors and by size of organizations
```

```python
df_ransomware.head(20)
```

```python
# Total number of ransom_paid paid or not or unknown per year

ransom_paid_per_year = df_ransomware.groupby('year')['ransom_paid'].value_counts().unstack(fill_value=0)
ransom_paid_per_year.plot(kind='bar', stacked=True)
plt.title('Ransom Paid Status per Year')
plt.ylabel('Count')
plt.xlabel('Year')
plt.show()
```

```python
#to have another vision, let's calculate the percentage of refused vs paid
ransom_paid_per_year = df_ransomware.groupby('year')['ransom_paid'].value_counts().unstack(fill_value=0)
percentages_per_year = ransom_paid_per_year.apply(lambda x: x / x.sum() * 100, axis=1)

# Round the percentages to 2 decimal places
percentages_per_year = percentages_per_year.round(2)

# Display the table
display(percentages_per_year)
```

```python
# Filter the DataFrame to include only 'paid ransom' entries
df_paid_ransoms = df_ransomware[df_ransomware['ransom_paid'] == 'ransom paid']

# Group by 'location' and count the occurrences
ransom_paid_per_country = df_paid_ransoms['location'].value_counts()

# Select the top 20 countries
top_20_countries = ransom_paid_per_country.head(20)

# Create a bar plot
plt.figure(figsize=(12, 6))
top_20_countries.plot(kind='bar')
plt.title('Number of Paid Ransoms per Country (Top 20)')
plt.xlabel('Country')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Number of paid ransom per sector

# Group by 'location' and count the occurrences
ransom_paid_per_sector = df_paid_ransoms['sector'].value_counts()

# Select the top 20 countries
top_20_sector = ransom_paid_per_sector.head(20)

# Create a bar plot
plt.figure(figsize=(12, 6))
top_20_sector.plot(kind='bar')
plt.title('Number of Paid Ransoms per Sectors (Top 20)')
plt.xlabel('Sector')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Group by 'location' and count the occurrences
ransom_paid_per_organisation_size = df_paid_ransoms['organisation_size'].value_counts()

# Create a bar plot
plt.figure(figsize=(12, 6))
ransom_paid_per_organisation_size.plot(kind='bar')
plt.title('Number of Paid Ransoms per Org size')
plt.xlabel('organisation size')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

plt.show()
```
