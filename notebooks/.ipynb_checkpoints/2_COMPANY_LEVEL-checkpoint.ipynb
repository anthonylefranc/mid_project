```python
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#df_ransomware = pd.read_csv('..\data\cleaned\clean_dataset.csv')
df_ransomware = pd.read_csv('clean_dataset.csv')
```

```python
#In this section, we'll be highlighting the sectors most affected, 
#and understanding how company size impacts their response to these attacks.

#We'll also try to understand whether one type of ransomware stands out according to company size or sector.
```

```python
df_ransomware.head()
```

```python
#Identificcation of the sectors
df_ransomware['sector'].unique()
```

```python
len(df_ransomware['sector'].unique())
```

```python
#Distribution of Ransomware by Sector

plt.figure(figsize=(10, 6)) 
sns.countplot(data=df_ransomware, x='sector')
plt.xticks(rotation=45) 
plt.title('Distribution of Ransomware by Sector')
plt.show()
```

```python
#Distribution of Ransomware by Organisation Size

#The 3 sectors which are the most targeted are:
#Government
#Healthcare
#Tech
#The medium and large companies are the msot target.

plt.figure(figsize=(10, 6))  
sns.countplot(data=df_ransomware, x='organisation_size')
plt.xticks(rotation=45)  
plt.title('Distribution of Ransomware by Organisation Size')
plt.show()
```

```python
num_ransom_paid = df_ransomware.loc[df_ransomware['ransom_paid'] == 'ransom paid', 'ransom_paid'].count()
num_refused = df_ransomware.loc[df_ransomware['ransom_paid'] == 'refused', 'ransom_paid'].count()
num_unknown = df_ransomware.loc[df_ransomware['ransom_paid'] == 'unknown', 'ransom_paid'].count()

print("Number of 'Ransom Paid':", num_ransom_paid)
print("Number of 'Refused':", num_refused)
print("Number of 'Unknown':", num_unknown)
```

```python
ransom_paid_by_sector = df_ransomware[df_ransomware['ransom_paid'].isin(['ransom paid', 'refused', 'unknown'])].groupby('sector')['ransom_paid'].value_counts()
print(ransom_paid_by_sector)
```

```python
#Distribution of Ransom Cases by Sector

ransom_counts = df_ransomware.groupby(['sector', 'ransom_paid']).size().unstack()

ransom_counts.plot(kind='bar', stacked=True, figsize=(10, 6))

plt.xlabel('Sector')
plt.ylabel('Number of Cases')
plt.title('Distribution of Ransom Cases by Sector')

plt.legend(title='Ransom Status')
plt.show()
```

```python
#Distribution of Ransom paid by Sector

ransom_paid_df = df_ransomware[df_ransomware['ransom_paid'] == 'ransom paid']

# Group by 'sector' and 'ransom_paid' and calculate the size
ransom_counts_paid = ransom_paid_df.groupby(['sector', 'ransom_paid']).size().unstack()

# Plot the data
ransom_counts_paid.plot(kind='bar', stacked=True, figsize=(10, 6))

plt.xlabel('Sector')
plt.ylabel('Number of Cases')
plt.title('Distribution of Ransom Cases (Paid) by Sector')

plt.legend(title='Ransom Status')

plt.show()
```

```python
#Distribution of Ransom refused by Sector

ransom_paid_df = df_ransomware[df_ransomware['ransom_paid'] == 'refused']

# Group by 'sector' and 'ransom_paid' and calculate the size
ransom_counts_paid = ransom_paid_df.groupby(['sector', 'ransom_paid']).size().unstack()

# Plot the data
ransom_counts_paid.plot(kind='bar', stacked=True, figsize=(10, 6))

plt.xlabel('Sector')
plt.ylabel('Number of Cases')
plt.title('Distribution of Ransom Cases (Paid) by Sector')

plt.legend(title='Ransom Status')

plt.show()
```

```python
#Distribution of Ransom Cases by O

ransom_counts = df_ransomware.groupby(['organisation_size', 'ransom_paid']).size().unstack()

ransom_counts.plot(kind='bar', stacked=True, figsize=(10, 6))

plt.xlabel('Sector')
plt.ylabel('Number of Cases')
plt.title('Distribution of Ransom Cases by organisation size')

plt.legend(title='Ransom Status')
plt.show()
```
