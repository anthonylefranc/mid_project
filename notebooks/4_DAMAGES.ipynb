```python
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#df_ransomware = pd.read_csv('..\data\cleaned\clean_dataset.csv')
df_ransomware = pd.read_csv('clean_dataset.csv')
```

```python
#In this section, we'll be highlighting the damages caused by this type of cyber-attacks
# and try to estimate the business impact (revenue loss)
```

```python
#Distribution of Ransomware payment
sns.countplot(x='ransom_paid', data=df_ransomware);
plt.title('Ransomware payment')
plt.xticks(rotation=90);
```

```python
#Mean Ransom cost by organisation size
mean_ransom_cost_by_size = df_ransomware.groupby('organisation_size')['ransom_cost'].mean()
print(mean_ransom_cost_by_size)
```

```python
#Mean Ransom Cost by Organization Size
#Cyber-criminals ask bigger amount of ransom to the large companies than the smallest
plt.figure(figsize=(8, 6))
mean_ransom_cost_by_size.plot(kind='bar')
plt.xlabel('Organization Size')
plt.ylabel('Mean Ransom Cost')
plt.title('Mean Ransom Cost by Organization Size')
plt.xticks(rotation=45)
plt.show()
```

```python
#Mean Ransom cost by sector
mean_ransom_cost_by_sector = df_ransomware.groupby('sector')['ransom_cost'].mean()
print(mean_ransom_cost_by_size)
```

```python
#Mean Ransom Cost by Sector 
plt.figure(figsize=(8, 6))
mean_ransom_cost_by_sector.plot(kind='bar')
plt.xlabel('Sector')
plt.ylabel('Mean Ransom Cost')
plt.title('Mean Ransom Cost by Sector')
plt.xticks(rotation=45)
plt.show()
```

```python
#Let's group the ransom cost by ranges
df_ransomware['ransom_cost'].unique()
```

```python
len(df_ransomware['ransom_cost'].unique())
```

```python
#Boundaries for the range groups
bins = [0, 1000000, 10000000, 100000000, float('inf')]

#Labels for the range groups
labels = ['0-1M', '1M-10M', '10M-100M', '100M+']

#New column with the range groups
df_ransomware['ransom_cost_group'] = pd.cut(df_ransomware['ransom_cost'], bins=bins, labels=labels, right=False)

#Updated DataFrame with the range groups
print(df_ransomware[['ransom_cost', 'ransom_cost_group']].head(10))
```

```python
df_ransomware['ransom_cost_group'].unique()
```

```python
len(df_ransomware['ransom_cost_group'].unique())
```

```python
# Group the data by ransom_cost and ransom_paid values, and count the occurrences
ransom_counts = df_ransomware.groupby(['ransom_cost_group', 'ransom_paid']).size().unstack()

# Create the plot
ransom_counts.plot(kind='bar', stacked=True, figsize=(10, 6))

# Set the labels and title of the plot
plt.xlabel('Ransom Cost')
plt.ylabel('Number of Cases')
plt.title('Number of Ransom Cases by Ransom Cost')

# Show the plot
plt.legend(title='Ransom Status')
plt.show()
```

```python
#Distribution of Ransom Cost by Sector

ransom_counts = df_ransomware.groupby(['sector', 'ransom_cost_group']).size().unstack()

ransom_counts.plot(kind='bar', stacked=True, figsize=(10, 6))

plt.xlabel('Sector')
plt.ylabel('Number of Cases')
plt.title('Distribution of Ransom Cases by Sector')

plt.legend(title='Ransom Status')
plt.show()
```

```python
#Distribution of Ransom Cost by Organisation size

ransom_counts = df_ransomware.groupby(['organisation_size', 'ransom_cost_group']).size().unstack()

ransom_counts.plot(kind='bar', stacked=True, figsize=(10, 6))

plt.xlabel('Sector')
plt.ylabel('Number of Cases')
plt.title('Distribution of Ransom Cases by Sector')

plt.legend(title='Ransom Status')
plt.show()
```

```python
#Number of ransomware groups
ransomware_counts = df_ransomware['ransomware'].value_counts()
print(ransomware_counts)
```

```python
#Top 10 of ransomware groups
ransomware_counts = df_ransomware['ransomware'].value_counts().drop(['unknown', 'not_revealed'])
top_10_ransomware = ransomware_counts.head(10)
print(top_10_ransomware)
```

```python
top_10_ransomware.plot(kind='bar', figsize=(10, 6))

plt.xlabel('Ransomware')
plt.ylabel('Occurrences')
plt.title('Top 10 Ransomware Occurrences')
plt.show()
```

```python
#Top 10 Ransomware Occurrences by Organisation size

top_10_ransomware_df = df_ransomware[df_ransomware['ransomware'].isin(top_10_ransomware.index)]

group_counts = top_10_ransomware_df.groupby(['organisation_size', 'ransomware']).size().unstack()

group_counts.plot(kind='bar', figsize=(12, 8))
plt.xlabel('Organisation size')
plt.ylabel('Occurrences')
plt.title('Top 10 Ransomware Occurrences by Sector')
plt.legend(title='Ransomware')
plt.show()
```

```python
#Top 5 Ransomware Occurrences by Ransom Cost Group

top_5_ransomware = top_10_ransomware.head(5)
top_5_ransomware_df = df_ransomware[df_ransomware['ransomware'].isin(top_5_ransomware.index)]
group_counts = top_5_ransomware_df.groupby(['ransom_cost_group', 'ransomware']).size().unstack()


group_counts.plot(kind='bar', figsize=(12, 8))
plt.xlabel('Ransom Cost Group')
plt.ylabel('Occurrences')
plt.title('Top 5 Ransomware Occurrences by Ransom Cost Group')
plt.legend(title='Ransomware')
plt.show()
```
